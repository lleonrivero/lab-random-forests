## 1. Import necessary libraries

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
import statsmodels.api as sm


import matplotlib.pyplot as plt
%matplotlib inline
```

## 2. Load datasets

```python
display(numerical.head(3))
display(numerical.shape)
display(categorical.head(3))
display(categorical.shape)
display(target.head(3))
display(target.shape)
```

## 3. Apply the Random Forests algorithm but this time only by upscaling the data.

```python
display(categorical.dtypes)

# The categorical dataset still contains a lot of columns that are actually numerical.
# We will be turning them into objects to OneHotEncode them at a later moment.

for i in categorical.columns:
    categorical[i]=categorical[i].astype('object')

display(categorical.dtypes)
```

```python
# Concatenating DataFrames
data = pd.concat([numerical, categorical, target], axis = 1)
display(data.shape)
data.columns
```

### 3.1 Xy split/Test-Train split

```python
y = data['TARGET_B']
X = data.drop(['TARGET_B', 'TARGET_D'], axis = 1) # TARGET_D????

numericalX = X.select_dtypes(np.number)
categoricalX = X.select_dtypes(np.object)

# we OneHotEncode the categoricals so we can use the same dataset to perform a regression later (in the lab).
# it is not needed for a DecisionTree or RandomForest model
# Notice this is not entirely correct: we are fitting our full dataset, not only the train set.
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop='first').fit(categoricalX)
encoded_categorical = encoder.transform(categoricalX).toarray()
encoded_categorical = pd.DataFrame(encoded_categorical)
X = pd.concat([numericalX, encoded_categorical], axis = 1)

# Note: we need to do train/test split before downsampling, and then only downsample the training set - Why?
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

### 3.2 Upscaling train set

```python
donate_upsample = pd.concat([X_train, y_train], axis = 1)
donate_upsample.shape
```

```python
donate_no = donate_upsample[donate_upsample['TARGET_B']==0]
donate_yes = donate_upsample[donate_upsample['TARGET_B']==1]
donate_yes_upsample = donate_yes.sample(len(donate_no), replace = True)
```

```python
print(donate_no.shape)
print(donate_yes_upsample.shape)
```

```python
donations = pd.concat([donate_no, donate_yes_upsample], axis=0)
X_train_upsample = donations.drop(['TARGET_B'], axis = 1)
y_train_upsample = donations['TARGET_B']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

clf = RandomForestClassifier(max_depth=5,
                             min_samples_split=20,
                             min_samples_leaf =20,
                             max_samples=0.2,
                             random_state = 42)
clf.fit(X_train_upsample, y_train_upsample)
print(clf.score(X_train_upsample, y_train_upsample))
print(clf.score(X_test, y_test))

y_pred = clf.predict(X_test)
display(y_test.value_counts())
display(confusion_matrix(y_test, y_pred))
```

## 4. Feature selection

```python
X_train.reset_index(drop = True)
X_test.reset_index(drop = True)
y_train.reset_index(drop = True)
y_test.reset_index(drop = True)
```

```python
X_test.shape
y_test.shape
```

```python
donations_data_a = pd.concat([X_train, y_train], axis = 1)
donations_data_b = pd.concat([X_test, y_test], axis = 1)
display(donations_data_a.shape)
display(donations_data_b.shape)
```

```python
donations_data = pd.concat([donations_data_a, donations_data_b], axis = 0).reset_index(drop=True)
donations_data.head()
```

```python
from sklearn.feature_selection import VarianceThreshold 
var_threshold = 0.02
sel = VarianceThreshold(threshold=(var_threshold))

# 1) default is threshold = 0 :eliminate all coumns that are identical for all rows
# 2) in practice we would scale the columns first, and then apply threshold, or apply 
#    a different threshold for different columns

# This drops the columns that have a variance less than this threshold
sel = sel.fit(donations_data)
temp = sel.transform(donations_data)
temp = pd.DataFrame(temp)
print(donations_data.shape)
print(temp.shape)
```

```python
pd.DataFrame(temp).head()
```
